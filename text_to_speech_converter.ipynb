```python
import tkinter as tk
from tkinter import *
from tkinter import filedialog,Frame
from tkinter.ttk import Combobox
import pyttsx3
import os
from gtts import gTTS
from playsound import playsound
```

```python
root = Tk()
root.title("Text to Speech Converter")
root.geometry("1000x580+200+80")
root.resizable(True,True)
root.configure(bg="#CCFFFF")
#root.mainloop()
```

```python
tts=pyttsx3.init()
def speaknow():
    text=text_box.get(1.0,END)
    gender=gender_box.get()
    speed=speed_box.get()
    voices=tts.getProperty('voices')
    
    def setvoice():
        if(gender=='Male'):
            tts.setProperty('voice',voices[0].id)
            tts.say(text)
            tts.runAndWait()
        else:
            tts.setProperty('voice',voices[1].id)
            tts.say(text)
            tts.runAndWait()
    if(text):
        if(speed=='Fast'):
            tts.setProperty('rate',250)
            setvoice()
        elif(speed=='Medium'):
            tts.setProperty('rate',150)
            setvoice()
        else:
            tts.setProperty('rate',60)
            setvoice()
            
```

```python
play_button=PhotoImage(file="C:/Users/KIIT/OneDrive/Desktop/button.png")
play_btn=Button(root,text="Play",compound=LEFT,image=play_button,bg="white",width=200,font="arial 14 bold",borderwidth='0.3c',command=speaknow)
play_btn.place(x=435,y=450)
root.mainloop()
```
